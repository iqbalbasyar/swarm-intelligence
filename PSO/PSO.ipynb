DIketahui sebuah fungsi :

$$f(x_1, x_2) = - \sum_{i=1}^{5} i \cos{((i+1)x_1 + 1)} \sum_{i=1}^{5} i \cos{((i+1)x_2 + 1)}$$

Tentukan nilai maksimum untuk fungsi tersebut

```python
import matplotlib.pyplot as plt
import math
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

def f(x,y):
    b2 = 0
    for i in range (1,6):
        b2 += np.cos((i+1)*y + 1)
        
    b1 = 0
    for i in range (1,6):
        b1 += np.cos((i+1)*x + 1)  * b2
    
    return b1
#plot


# Make data.
X = np.arange(-50, 50, 0.25)
Y = np.arange(-50, 50, 0.25)
X, Y = np.meshgrid(X, Y)
Z = f(X,Y)

fig = plt.figure()
ax = fig.gca(projection='3d')

surf = ax.plot_surface(X, Y, Z)
plt.show()
```
